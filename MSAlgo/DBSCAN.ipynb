```python
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
random.seed(a=1)
```

```python
num_row, num_col = 17,17
```

```python
example = 5
samples = []

while example:
    samples = []
    for i in range(2000):
        x=random.randrange(0,num_row-1)
        y=random.randrange(0,num_col-1)
        samples += [[x,y]]
    x,y = list(zip(*samples))
    sns.kdeplot(x=x, y=y, cmap="Reds", shade=True, bw_adjust=.5)
    plt.show()
    example -= 1
```

```python
_ = 5
batch = []

while _:
    samples = []
    for i in range(1000):
        x=random.randrange(0,num_row-1)
        y=random.randrange(0,num_col-1)
        samples += [[x,y]]
    x,y = list(zip(*samples))
    batch += [pd.DataFrame({"x": x, "y": y})]
    _ -=1
```
